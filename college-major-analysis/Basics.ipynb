```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import display

%matplotlib inline
```

# Analyzing Post-College Employment By Major

What information and conclusions can be derived from a dataset (collected and released by <a href="https://www.census.gov/programs-surveys/acs/">American Community Survey</a>) of college graduates from the years 2010-2012.

## Read and clean the data

```python
grads = pd.read_csv("recent-grads.csv")
```

```python
print("Table shape: {}".format(grads.shape))
print("Sample top:")
display(grads.head())
print("Sample bottom:")
display(grads.tail())
print("Table description:")
display(grads.describe(include="all"))
```

#### Note:
To better accomodate <code>matplotlib</code>, any rows with NaN values must be dropped:

```python
before = grads.shape
after = grads.dropna().shape

print("{} -> {}".format(before, after))
```

```python
grads.dropna(inplace=True)
```

## Plotting the data:

### Scatter plots

```python
fig = plt.figure(figsize=(14, 15))
ax11 = fig.add_subplot(321)
ax12 = fig.add_subplot(322)
ax21 = fig.add_subplot(323)
ax22 = fig.add_subplot(324)
ax31 = fig.add_subplot(325)
ax32 = fig.add_subplot(326)

grads.plot(x="Sample_size", y="Median", 
           kind="scatter", 
           ax=ax11,
           title="Sample Size vs Median")

grads.plot(x="Sample_size", y="Unemployment_rate", 
           kind="scatter", 
           ax=ax12,
           title="Sample Size vs Unemployment Rate")

grads.plot(x="Full_time", 
           y="Median", 
           kind="scatter", 
           ax=ax21,
           title="Full Time vs Median")

grads.plot(x="ShareWomen", 
           y="Unemployment_rate", 
           kind="scatter", 
           ax=ax22,
           title="Share Women vs Unemployment Rate")

grads.plot(x="Men", 
           y="Median", 
           kind="scatter", 
           ax=ax31,
           title="Men vs Median")

grads.plot(x="Women", 
           y="Median", 
           kind="scatter", 
           ax=ax32,
           title="Women vs Median")
```

### Scatter matrices

```python
from pandas.plotting import scatter_matrix as sm
```

```python
sm(grads[["Sample_size", "Median"]], figsize=(12, 12))
```

```python
sm(grads[["Sample_size", "Median", "Unemployment_rate"]], figsize=(15,15))
```

#### Note:
First, it was clear from the beginning that the sample sizes were way to small on average. The scatter plots are mostly blank space as the axes takes into account the few large sample sizes, accordingly skewed towards the margins.

Second, it would be interesting to compare this data to the average in a country/region. The histograms show tendencies, but they don't have much meaning alone.

Now for some observations about the data itself:

* There is no correlation between the employment rate and median salary, as there shouldn't be

* Scatter plots of sample size to X, seem to be the equivalent of showing the histogram of X

Let's look at more meaningful correlations:

```python
sm(grads[["Total", "Median", "Unemployment_rate"]], figsize=(15,15))
```

```python
sm(grads[["Men", "Low_wage_jobs", "Full_time"]], figsize=(15,15))
```
